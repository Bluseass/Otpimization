Этот код сначала определяет функцию f(x), затем реализует метод золотого сечения в функции golden_section_search. Результатом является значение экстремума функции и его координата. Затем код строит график функции с помощью библиотеки matplotlib и отмечает найденный экстремум на графике.

Лабораторная работа ИСО № 3 - Метод Ньютона-Рафсона

Для решения данной задачи с использованием метода Ньютона-Рафсона, нам сначала необходимо найти частные производные функции y=(x-3)^2+(y-4)^2 по переменным x и y.

dy/dx = 2(x-3)  
dy/dy = 2(y-4)  

Затем мы можем использовать итерационный метод Ньютона-Рафсона для нахождения экстремума функции. Рекурсивная формула для метода Ньютона-Рафсона выглядит следующим образом:

x_new = x_old - f(x_old)/f'(x_old)

где x_old - предыдущее значение переменной x, x_new - новое значение переменной x, f(x) - функция, для которой мы ищем экстремум, и f'(x) - ее производная.

В данной задаче мы будем искать минимум функции, поэтому при расчете нового значения переменной, мы будем вычитать значение функции, а не прибавлять.

Реализация задачи на языке Python с использованием библиотеки matplotlib для визуализации выглядит следующим образом:

```python
import pandas as pd
df = pd.DataFrame(columns = ['x','y'])
x1 =[]
x2 = []
import numpy as np
import matplotlib.pyplot as plt

def newton_raphson(x, y):
    # Функция, для которой ищем экстремум
    f = (x-3)**2 + (y-4)**2
    # Частные производные функции
    dfdx = 2*(x-3)
    dfdy = 2*(y-4)
    # Рекурсивная формула метода Ньютона-Рафсона
    x_new = x - f/dfdx
    y_new = y - f/dfdy

    return x_new, y_new

def plot_gradient_descent(x_history, y_history):
    x = np.linspace(-10, 10, 100)
    y = np.linspace(-10, 10, 100)
    X, Y = np.meshgrid(x, y)
    Z = (X-3)**2 + (Y-4)**2

    fig, ax = plt.subplots(figsize=(8, 6))
    ax.contour(X, Y, Z, levels=20)
    ax.scatter(x_history, y_history, color='red', marker='o', label='Iterations')
    ax.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('минимализация (x-3)^2 + (y-4)^2 ')

    plt.show()

def newton_raphson_minimization(x_init, y_init):
    x_history = [x_init]
    y_history = [y_init]
    x = x_init
    y = y_init
    while True:

        x_new, y_new = newton_raphson(x, y)
        x_history.append(x_new)
        y_history.append(y_new)
        # Остановка итераций, когда изменение значений становится меньше 0.01
        if abs(x_new - x) < 0.01 and abs(y_new - y) < 0.01:
            break
        x = x_new
        y = y_new

    return x_history, y_history

# Начальные значения переменных
x_init = 0
y_init = 0

# Минимизация функции с использованием метода Ньютона-Рафсона
x_history, y_history = newton_raphson_minimization(x_init, y_init)
df['y'] = y_history
df['x'] = x_history
# Вывод истории изменения переменных при итерациях
for i in range(len(x_history)):
    print(f'Iteration {i+1}: x = {x_history[i]:.2f}, y = {y_history[i]:.2f}')
plot_gradient_descent(x_history, y_history)
# Визуализация изменения переменных и контурн
df
```

Лабораторная работа ИСО № 4
Тема: Методы условной оптимизации
Метод барьерных поверхностей (метод внутренней точки)
Ставим барьеры и расширяем их при приблежении, похоже на метод штрафных функций, но тут ограничения могут быть только в виде неравенств. Поиск ведется внутри барьера и находится минимум.

```python

import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt

# Определяем функцию
def f(params):
    x, y = params
    return x + 4*y + x*y - 2*x**2 - 2*y**2

# Определяем ограничения
def constraint1(params):
    x, y = params
    return 12 - x - 2*y

def constraint2(params):
    x, y = params
    return 15 - 3*x - y

constraints = [{'type': 'ineq', 'fun': constraint1},
               {'type': 'ineq', 'fun': constraint2}]

# Определяем функцию барьерной поверхности
def barrier_function(params, r):
    return f(params) - r * (np.log(-constraint1(params)) + np.log(-constraint2(params)))

# Определяем градиент барьерной поверхности
def barrier_gradient(params, r):
    x, y = params
    grad1 = np.array([2*x**2 - y + 12 - x, 4 - 2*y + 2*x])
    grad2 = np.array([6 - 2*x - y, 4 - 2*x + 2*y])
    return grad1 / constraint1(params) + grad2 / constraint2(params)

# Определяем начальную точку и параметры метода
x0 = np.array([0, 1])
r = 10
alpha = 0.1

# Итеративно находим экстремум
while r > 1e-5:
    minimize_result = minimize(barrier_function, x0, args=(r,), method='SLSQP', jac=barrier_gradient, constraints=constraints)
    x0 = minimize_result.x
    r *= alpha

# Выводим найденное решение
print("x =", x0[0])
print("y =", x0[1])
print("f(x, y) =", f(x0))

# Отображаем график функции
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z = f([X, Y])
plt.contourf(X, Y, Z, levels=100, cmap='jet')
plt.colorbar()
plt.scatter(x0[0], x0[1], color='red', label='Solution')
plt.xlabel('x')
plt.ylabel('y')
plt.title('f(x, y)')
plt.legend()
plt.show()

```

Лабораторная работа №5 -
Метод стохастической аппроксимации
При выполнении данного кода будет выведен результат поиска экстремума функции и построен график движения на каждой итерации. Результатом будет точка экстремума `[x_res, y_res]` и значение функции в этой точке `f_res`.

Метод итерационный (существует шаг сходимости), может подстраиваться под динамическую систему. Т.к чувствительность экспотенциально устойчива т.к используется квадратическая аппроксимация, это позволит представить плотность в форме нормального распределения с ковариационнолй матрицей e^2W. Что характерезует разброс вокруг системы равновесия !x .

Плюсы модели:
1.описывает любую тензенцию
2.писывает многофакторные зависимости
3.модель отсеивает шумы/подстраивает и модернизирует переменных/адаптируется к изменениям тенденций.
Минусы:
1.Модель громозтка.
2.Нет никакой формулы или правила по которым можно взять начальные коэфиценты (они случайны) и неудачная конфигурация в теории может замедлить решение задачи.

```python

import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return (x - 3)**2 + (y - 4)**2

def approximate_extremum(f, start_point, accuracy, step):
    x, y = start_point
    x_vals = [x]
    y_vals = [y]

    while True:
        delta_x = np.random.uniform(-step, step)
        delta_y = np.random.uniform(-step, step)

        new_x = x + delta_x
        new_y = y + delta_y

        if f(new_x, new_y) < f(x, y):
            x = new_x
            y = new_y
            x_vals.append(x)
            y_vals.append(y)

        if abs(f(x, y) - f(x_vals[-1], y_vals[-1])) < accuracy:
            break

    return x, y, f(x, y), x_vals, y_vals

# Начальная точка
start_point = [0, 3]

# Точность
accuracy = 0.01

# Шаг
step = 0.2

# Вызываем функцию аппроксимации экстремума
x_res, y_res, f_res, x_vals, y_vals = approximate_extremum(f, start_point, accuracy, step)

print("Результат:")
print("x =", x_res)
print("y =", y_res)
print("f(x, y) =", f_res)

# График поиска экстремума
plt.plot(x_vals, y_vals, 'bo-')
plt.plot(x_res, y_res, 'ro')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Поиск экстремума функции")
plt.grid()
plt.show()

```
